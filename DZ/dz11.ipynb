```python
from sympy import symbols, Interval, diff
from sympy import *
from sympy.plotting import plot
from sympy.calculus.util import minimum, maximum
import math
```

```python
x = symbols('x', real = True)
f = x**2 - abs(x)*6+8
f
```

```python
sol_res = solveset(f, x, Interval(-oo, oo))
result = []
for i in sol_res:
    result.append(i)
print(f'Корни функции {f} = {result}')
```

```python
plot(f,(x, -5, 5))
```

```python
d_f = diff(f)
print(f'Производная функции y = {d_f}')
```

```python
exstremum = solveset(d_f, x, Interval(-oo, oo))
exst_list = []
for i in exstremum:
    exst_list.append(i)
print(f'Экстремумы функции {exst_list}')
```

```python
plot(d_f,f, (x,-5, 5))
```

```python
print(f'Интервалы на которых функция возрастает или падает:') 
for i in range(0,len(exst_list)):
    tochka = exst_list[i]
    if i == 0: 
        if d_f.subs(x,tochka-0.000001) < 0: print(f'Функция на промежутке {-oo}:{tochka} падающая')
        else: print(f'Функция на промежутке {-oo}:{tochka} возрастает') 
    elif i == tochka-1:
        if d_f.subs(x,tochka+0.000001) < 0: print(f'Функция на промежутке {tochka}:{oo} падающая')
        else: print(f'Функция на промежутке {tochka}:{oo}: возрастает')
    else:
        if d_f.subs(x,tochka-0.000001) < 0: print(f'Функция на промежутке {exst_list[i-1]}:{tochka} падающая')
        elif d_f.subs(x,tochka-0.000001) > 0: print(f'Функция на промежутке {exst_list[i-1]}:{tochka}: возрастает')
            
        if d_f.subs(x,tochka+0.000001) < 0: print(f'Функция на промежутке {tochka}:{exst_list[i+1]} падающая')
        elif d_f.subs(x,tochka+0.000001) > 0: print(f'Функция на промежутке {tochka}:{exst_list[i+1]} возрастает')
```

```python
print(f'Интервалы на которых функция больше или меньше нуля:') 
for i in range(0,len(result)):
    tochka = result[i]
    if i == 0: 
        if f.subs(x,tochka-0.000001) < 0: print(f'Функция на промежутке {-oo}:{tochka} меньше 0')
        else: print(f'Функция на промежутке {-oo}:{tochka} больше 0') 
    elif i == tochka-1:
        if f.subs(x,tochka+0.000001) < 0: print(f'Функция на промежутке {tochka}:{oo} меньше 0')
        else: print(f'Функция на промежутке {tochka}:{oo}: больше 0')
    else:
        if f.subs(x,tochka-0.000001) < 0: print(f'Функция на промежутке {result[i-1]}:{tochka} меньше 0')
        elif f.subs(x,tochka-0.000001) > 0: print(f'Функция на промежутке {result[i-1]}:{tochka}: больше 0')
            
        if f.subs(x,tochka+0.000001) < 0: print(f'Функция на промежутке {tochka}:{result[i+1]} меньше 0')
        elif f.subs(x,tochka+0.000001) > 0: print(f'Функция на промежутке {tochka}:{result[i+1]} больше 0')
```
