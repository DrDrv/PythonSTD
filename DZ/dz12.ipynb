```python
import os
import pandas as pd
import matplotlib.pyplot as plt

root = os.getcwd()
filename = os.path.join(root,'csd.csv')
data = pd.read_csv(filename)
data#.head()
```

```python
head_list = ['month_number','facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer','total_units','total_profit']
k = 0
for i in head_list:
    head_list[k] = [i,data[i].tolist()]
    k += 1
```

```python

for i in range(1,k-2):
    plt.plot((head_list[0])[1], (head_list[i])[1],linewidth=3, label=(head_list[i])[0], marker='o' )
plt.legend(loc='upper left')
plt.title('Прибыль по месяцам')
plt.xlabel('месяцы')
plt.ylabel('Доходы')
plt.xticks((head_list[0])[1])
plt.show
```

```python
bathingsoap = data['bathingsoap'].tolist()
plt.grid(linewidth=1, linestyle='--')
plt.bar((head_list[0])[1],bathingsoap,label = 'bathingsoap', align = 'center')
plt.legend(loc='upper left')
plt.title('продажи мыла для купания')
plt.xlabel('месяцы')
plt.ylabel('Продажи')
plt.xticks((head_list[0])[1])

plt.show
```

```python
labels = [ (head_list[i])[0] for i in range(1,len(head_list)-2)]
sizes = [ sum((head_list[i])[1]) for i in range(1,len(head_list)-2)]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()
```
